```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Perceptron
```

```python
df = pd.read_csv(r"G:\Python Projects,Docs\Machine learning\DL\titanic.csv")
```

```python
df
```

```python
df1=df.drop(columns=["Name","Sex","Ticket","Cabin","Embarked"])
```

```python
df1.corrwith(df["Survived"])
```

```python
df2 = df1.drop(columns=["PassengerId","Age","Parch","SibSp"])
```

```python
x = df2.drop(columns="Survived")
y = df2.Survived
```

```python
x=x.fillna(0)
y
```

```python
import seaborn as sns
sns.scatterplot(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
p_model = Perceptron()
```

```python
p_model.fit(x_train,y_train)
```

```python
p_model.coef_    #the weights of the inputs in the perceptron 
```

```python
p_model.intercept_    #the bias of the perceptron
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(p_model.predict(x_test),y_test)
```
